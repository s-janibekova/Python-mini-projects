```python
from collections import defaultdict
doctors = {'schedule':{}}

def schedule_for_doctor():
    d1 = datetime.date(2019,1,1)
    d2 = datetime.date(2020, 1, 1)
    days = [d1 + datetime.timedelta(days=x) for x in range((d2-d1).days + 1)]
    str_date = []
    for day in days:
        str_date.append(day.strftime('%Y%m%d'))
    caldendar_of_doctor_for_year = {k:0 for k in str_date}
    doctors['schedule'].update(caldendar_of_doctor_for_year)
    return doctors

schedule_for_doctor()
```
